```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
```

# **Dataset Iris**

```
dataset = pd.read_csv('/content/drive/My Drive/BCML/datasets/iris.csv')

dataset = dataset.drop(['Id', 'Species'], axis=1)
```

```
dataset.head()
```

```
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=10)
    kmeans.fit(dataset)
    wcss.append(kmeans.inertia_) # Within cluster sum of squares

print(wcss)
```

```
plt.plot(range(1, 11), wcss)
plt.title('Elbow method')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
```

```
kmeans = KMeans(n_clusters=3, random_state=10)
model = kmeans.fit(dataset)
```

```
model.labels_
```

```
model.cluster_centers_
```

```
prediction = kmeans.predict(dataset)
```

## **SepalLengthCm-SepalWidthCm**

```
plt.scatter(dataset.values[prediction == 0, 0], dataset.values[prediction == 0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(dataset.values[prediction == 1, 0], dataset.values[prediction == 1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[prediction == 2, 0], dataset.values[prediction == 2, 1], s=100, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=50, c='yellow', label = 'Centroids')
```

## **SepalLengthCm-PetalLengthCm**

```
plt.scatter(dataset.values[prediction == 0, 0], dataset.values[prediction == 0, 2], s=100, c='red', label ='Cluster 1')
plt.scatter(dataset.values[prediction == 1, 0], dataset.values[prediction == 1, 2], s=100, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[prediction == 2, 0], dataset.values[prediction == 2, 2], s=100, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 2], s=50, c='yellow', label = 'Centroids')
```

## **SepalWidthCm-PetalWidthCm**

```
plt.scatter(dataset.values[prediction == 0, 1], dataset.values[prediction == 0, 3], s=100, c='red', label ='Cluster 1')
plt.scatter(dataset.values[prediction == 1, 1], dataset.values[prediction == 1, 3], s=100, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[prediction == 2, 1], dataset.values[prediction == 2, 3], s=100, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 1], model.cluster_centers_[:, 3], s=50, c='yellow', label = 'Centroids')
```

```
plt.scatter(dataset.values[prediction == 0, 0], dataset.values[prediction == 0, 3], s=100, c='red', label ='Cluster 1')
plt.scatter(dataset.values[prediction == 1, 0], dataset.values[prediction == 1, 3], s=100, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[prediction == 2, 0], dataset.values[prediction == 2, 3], s=100, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 3], s=50, c='yellow', label = 'Centroids')
```

# **Dataset Customer**

```
dataset = pd.read_csv('/content/drive/My Drive/BCML/datasets/customer.csv')
```

```
dataset = dataset.drop(['CustomerID'], axis=1)
```

```
dataset.head()
```

```
genre_categorized=pd.get_dummies(dataset['Genre'])
genre_categorized
```

```
dataset = pd.concat([dataset,genre_categorized],axis=1)
dataset.head()
```

```
dataset.describe()
```

```
dataset = dataset.drop(['Genre'], axis=1)
dataset.head()
```

```

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=10)
    kmeans.fit(dataset)
    wcss.append(kmeans.inertia_) # Within cluster sum of squares

print(wcss)
```

```
plt.plot(range(1, 11), wcss)
plt.title('Elbow method')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
```

## **n_clusters=6**
